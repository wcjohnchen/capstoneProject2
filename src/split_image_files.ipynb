```python
##### import
import os
import shutil

import numpy as np
import pandas as pd
```

```python
def pricing(x):
    if x >= 0 and x < 500000:
        return "low"
    elif x >= 500000 and x < 1000000:
        return "medium"
    else: ### x >= 1000000
        return "high"
```

```python
##### dataset
modified = pd.read_csv("../data/modified_data.csv")
modified.head()
```

```python
##### copy dataset
data = modified.copy()
#data.shape
```

```python
##### set price range
data['price_range'] = data['price'].apply(pricing)
data
```

```python
#print(len(data[data['price_range']=='low']))
#print(len(data[data['price_range']=='medium']))
#print(len(data[data['price_range']=='high']))
```

```python
##### set labels
data['label'] = data['price_range']
data['label'] = data['label'].apply(lambda x: 0 if x=='low' else x)
data['label'] = data['label'].apply(lambda x: 1 if x=='medium' else x)
data['label'] = data['label'].apply(lambda x: 2 if x=='high' else x)
data
```

```python
allFileNames = os.listdir('../image')
np.random.seed(10)
np.random.shuffle(allFileNames)
#print(allFileNames)
val_ratio = 0.1
test_ratio = 0.2

val_FileNames, test_FileNames, train_FileNames = np.split(np.array(allFileNames), [int(len(allFileNames)*val_ratio), int(len(allFileNames)*val_ratio + len(allFileNames)*test_ratio)])
#print(len(train_FileNames))
#print(len(val_FileNames))
#print(len(test_FileNames))
```

```python
#print(len(data[data['label']==0]))
#print(len(data[data['label']==1]))
#print(len(data[data['label']==2]))
```

```python
train=[]
for i in range(len(train_FileNames)):
    t = int(train_FileNames[i][:-4])
    train.append(int(t))
#print(train_FileNames)
#print(train)

test=[]
for i in range(len(test_FileNames)):
    t = test_FileNames[i][:-4]
    test.append(int(t))
#print(test_FileNames)
#print(test)

validation=[]
for i in range(len(val_FileNames)):
    t = val_FileNames[i][:-4]
    validation.append(int(t))
#print(val_FileNames)
#print(validation)
#print(dependent)
```

```python
##### move images to train, validation, and test folders
directory_ = r'../image'
for filename in os.listdir(directory_):
    index = int(filename[:-4])

    if filename.endswith(".jpg") and filename in train_FileNames:
        if index in train:
            if dependent['image_id'][index] == index and dependent['label'][index] == 0:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/train/class0/", filename))
            elif dependent['image_id'][index] == index and dependent['label'][index] == 1:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/train/class1/", filename))            
            elif dependent['image_id'][index] == index and dependent['label'][index] == 2:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/train/class2/", filename))     
                
    elif filename.endswith(".jpg") and filename in val_FileNames:
        if index in validation:
            if dependent['image_id'][index] == index and dependent['label'][index] == 0:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/validation/class0/", filename))
            elif dependent['image_id'][index] == index and dependent['label'][index] == 1:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/validation/class1/", filename))            
            elif dependent['image_id'][index] == index and dependent['label'][index] == 2:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/validation/class2/", filename))  

    elif filename.endswith(".jpg") and filename in test_FileNames:
        if index in test:
            if dependent['image_id'][index] == index and dependent['label'][index] == 0:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/test/class0/", filename))
            elif dependent['image_id'][index] == index and dependent['label'][index] == 1:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/test/class1/", filename))            
            elif dependent['image_id'][index] == index and dependent['label'][index] == 2:
                shutil.copy(os.path.join("../image/", filename), os.path.join("../image2/test/class2/", filename))
```
